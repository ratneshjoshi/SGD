```python
with open('gendered_words.txt', 'r') as infile:
    data = infile.read()
```

```python
male_words = []
female_words = []

for word_pairs in data.split('\n'):
    male_words.append(word_pairs.split('\t')[0])
    female_words.append(word_pairs.split('\t')[1])
```

```python
import pandas as pd
```

```python
df_train = pd.read_csv('data/train.csv')

df_train.columns
```

```python
training_data = df_train['text'].to_list()
training_data_debiased = []

i = 0
for instance in training_data:
    instance_debiased = []
    if i==12:
        print(instance))

    for x in instance.split(' '):
        if x not in male_words or x not in female_words:
            if i==12:
                print(x)
            instance_debiased.append(x)
        elif x in male_words:
            xx = female_words[male_words.index(x)]
            instance_debiased.append(xx)
            if i==12:
                print(x)
                print('poop', x, xx)
        elif x in female_words:
            xx = male_words[female_words.index(x)]
            instance_debiased.append(xx)
            if i==12:
                print(x)
                print('pp', x, xx)
        else:
            print(f"odd input {x}")
    
    instance_debiased = " ".join(instance_debiased)

    i = i+1

    if i==13:
        print(instance)
        print(instance_debiased)
        break


    training_data_debiased.append(instance_debiased)
```

```python
df_train_debiased = pd.DataFrame.from_dict({'text': training_data_debiased})

df_train_debiased.to_csv('data/train_debiased.csv')
```
